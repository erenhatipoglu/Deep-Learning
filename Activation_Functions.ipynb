```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Defining activation functions mathematically
# Sigmoid Function
def sigmoid(x):
    result = []
    for i in x:
        result.append(1 / (1 + math.exp(-i)))
    return result
```

```python
# Hyperbolic Tangent Function
def tanh(x, derivative=False):
    if derivative:
        return 1 - (np.tanh(x) ** 2)
    return np.tanh(x)
```

```python
# ReLU Function
def relu(x):
    result = []
    for i in x:
        result.append(max(0, i))
    return result
```

```python

# Leaky ReLU Function
def leaky_relu(x, alpha=0.1):
    result = []
    for i in x:
        result.append(i if i >= 0 else alpha * i)
    return result
```

```python
# Define the range for the graph
x = np.linspace(-3., 3., 100)
sig = sigmoid(x)
tanh_values = tanh(x)
relu_values = relu(x)
leaky_relu_values = leaky_relu(x)
swish = [s * s_val for s, s_val in zip(sig, x)]
```

```python
# Plotting the functions and displaying the graph
plt.plot(x, sig, label='Sigmoid')
plt.plot(x, tanh_values, label='Tanh')
plt.plot(x, relu_values, label='ReLU')
plt.plot(x, leaky_relu_values, label='Leaky ReLU')
plt.plot(x, swish, label='Swish')

# Adding labels and legend
plt.legend()
plt.axhline(y=0, color='k', linestyle='--')
plt.axvline(x=0, color='k', linestyle='--')
plt.title("Activation Functions")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()
```
